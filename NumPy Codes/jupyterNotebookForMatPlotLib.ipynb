```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = ['part1','part2','part3','part4','part5','part6','part7']
y = [80,50,20,90,80,95,65]

# Note - A bar plot in Matplotlib is a powerful visualization tool used to represent categorical data with rectangular bars, where the length of each bar is proportional to the value it represents. 
# Bar plots are useful for comparing different groups or showing changes over time.
plt.bar(x,y)
plt.xlabel('Harry Potter movies parts')
plt.ylabel('Popularity')
plt.title('Popularity of different parts of harry potter series')

# Note - Use plt.bar() for vertical bar plots and plt.barh() for horizontal bar plots, we can also Change colors, add error bars, and set widths.
```

```python
data = pd.read_excel('expense3.xlsx')
df = pd.DataFrame(data)
print(df)
```

```python
plt.bar(df["Payment Mode"],df["Amount"])
plt.show()
```

```python
grouped_by = df.groupby('Payment Mode')['Amount'].sum()

plt.bar(grouped_by.index,grouped_by.values)
```

```python
# Line Plot
# Note - Creating a line plot in Matplotlib is a great way to visualize trends over time or continuous data.
```

```python
x = ['day1','day2','day3','day4','day5','day6','day7']
y = [80,150,200,90,180,195,65]
y1 = [100,40,100,190,110,95,165]

plt.plot(x,y,marker='o',ls='--',color='green',label='week1')
plt.plot(x,y1,marker='d',ls='-',color='red',label='week2')
plt.legend()
plt.show()

# Note -
# plt.show(): Displays the plot window and is essential for visualizing plots in scripts.
# plt.legend(): Adds a legend to the plot for better understanding, especially when multiple datasets are involved.
```

```python
# Scatter Plot: it is Effective for visualizing relationships and distributions between two variables.

x = np.random.randint(1,10,50)
y = np.random.randint(10,100,50)
color = np.random.randint(10,100,50)
size = np.random.randint(10,100,50)

plt.scatter(x,y,marker='*',c=color,s=size)
plt.show()
```

```python
# Pie chart - A pie chart is a circular statistical graphic that is divided into slices to illustrate numerical proportions.
# Useful for visualizing proportions and categorical data.

# Autopct: Display percentages on the pie slices.
# Exploding Slices: Emphasize a particular slice by "exploding" it out from the center.
```

```python
brands = ['Oppo','Vivo','Apple','Samsung','OnePlus']
sale = [22,30,35,3,10]
c = ['red','pink','green','blue','yellow']
ex = [0.1,0,0,0,0]

plt.pie(sale,labels=brands,autopct='%.2f',explode=ex,shadow=True,colors=c)
plt.show()
```

```python
grouped_by = df.groupby('Payment Mode')['Amount'].sum()
plt.pie(grouped_by.values,labels=grouped_by.index,autopct='%.2f')
plt.show()
```
